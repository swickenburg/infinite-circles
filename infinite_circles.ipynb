```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## create inverted circles

```python
y_grid_size = 5
x_grid_size = 10

# vertical lines
y_line_dict = {}
for line_y in np.arange(y_grid_size + 1):
    line_y = line_y + 1
    x = np.arange(-1000, 1000,0.01)
    r = np.sqrt(x**2 + line_y**2)
    theta = np.arctan(x/line_y)
    x_plot = - 1/r * np.cos(theta)
    y_plot = 1/r * np.sin(theta)
    y_line_dict[line_y] = {}
    y_line_dict[line_y]['x'] = x_plot
    y_line_dict[line_y]['y'] = y_plot

# horizontal lines
x_line_dict = {}
for line_x in np.arange(-x_grid_size,x_grid_size):
    line_x = line_x + 0.5
    y = np.arange(0, 1000,0.01)
    r = np.sqrt(line_x**2 + (y + 1)**2)
    
    theta = np.arctan(line_x/(y + 1))
    x_plot = - 1/r * np.cos(theta)
    y_plot = 1/r * np.sin(theta)
    
    x_line_dict[line_x] = {}
    x_line_dict[line_x]['x'] = x_plot
    x_line_dict[line_x]['y'] = y_plot

# circles
circle_dict = {}
for circle_x in np.arange(-x_grid_size,x_grid_size):
    circle_dict[circle_x] = {}
    for circle_y in np.arange(y_grid_size):
        x_center = circle_x + 0.5
        y_center = circle_y + 1.5
        
        theta_circle = np.linspace(0,2*np.pi,1000)
        
        x = 0.5 * np.cos(theta_circle) + x_center
        y = 0.5 * np.sin(theta_circle) + y_center
        
        r = np.sqrt(x**2 + y**2)
        theta = np.arctan(x/y)
        x_plot = - 1/r * np.cos(theta)
        y_plot = 1/r * np.sin(theta)
        
        circle_dict[circle_x][circle_y] = {}
        circle_dict[circle_x][circle_y]['x'] = x_plot
        circle_dict[circle_x][circle_y]['y'] = y_plot
        
```

```python
def plot_circles(y_line_dict, x_line_dict, circle_dict, file_name='circles'):
    fig, ax = plt.subplots(1,1,figsize=(10,10))
    for i,line_y in enumerate(y_line_dict.keys()):
        x_plot = y_line_dict[line_y]['x']
        y_plot = y_line_dict[line_y]['y']

        ax.plot(x_plot,y_plot,c=f'C{i}',lw=2)
        ax.grid()

    for line_x in x_line_dict.keys():
        x_plot = x_line_dict[line_x]['x']
        y_plot = x_line_dict[line_x]['y']
        ax.plot(x_plot,y_plot,'darkgrey',lw=2)
        ax.grid()

    for circle_x in circle_dict.keys():
        for i,circle_y in enumerate(circle_dict[circle_x].keys()):
            x_plot = circle_dict[circle_x][circle_y]['x']
            y_plot = circle_dict[circle_x][circle_y]['y']
            ax.plot(x_plot,y_plot,c=f'C{i}',lw=2)
            ax.grid()

    ax.set_aspect(1)
    plt.savefig(f'{file_name}.png',bbbox_inches='tight',dpi=150)
```

```python
plot_circles(y_line_dict, x_line_dict, circle_dict)
```
